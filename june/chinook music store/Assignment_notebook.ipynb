### Question1: which city has the best customers?

### Question2: which country have the most invoices?
i.e the total number of invoices per country

```python
most_invoices = invoice.groupby('BillingCountry')['Total'].count().sort_values(ascending=False).reset_index()
most_invoices.head()
```

### Question3: who is the best customer?
i.e customer that spent the most

```python
#merge customer and invoice total on customerid to know customers spending
customer_invoice = pd.merge(customer, invoice[['CustomerId', 'Total']], on = 'CustomerId', how = 'inner')  
```

```python
customer_invoice.tail(3)
```

```python
#arrange in descending order to know the customer with highest spending
customer_invoice.groupby('CustomerId')['Total'].sum().sort_values(ascending= False).head(1)
```

### Question4: return the email, firstname and lastname of rock music listeners and arrange alphabetically

```python
#merge the invoice and invoice line data together
invoice_merge = pd.merge(invoice[['CustomerId', 'InvoiceId', 'BillingCountry']], invoice_line, on= 'InvoiceId', how='outer') 
```

```python
invoice_merge.tail(3)
```

```python
#merge the tracks and genre data together
track_merge = pd.merge(genre, track[['TrackId', 'GenreId']], on = 'GenreId', how='left')  
```

```python
track_merge.tail(3)
```

```python
#merge the invoice,invoice line, track and genre data alltogether
genre_merge = pd.merge(track_merge, invoice_merge, on = 'TrackId') 
```

```python
genre_merge.tail(3)
```

```python
#specify for only those that listen to rock music
rock = genre_merge[genre_merge['Name'] == 'Rock'] 
```

```python
#merge rock listeners with customer data to get customers name and mail
rock_customers = pd.merge(customer, rock, on = 'CustomerId') 
```

```python
rock_customers.tail(3)
```

```python
#rock music customers name and mail
rock_list = rock_customers[['FirstName', 'LastName', 'Email', 'Name']].sort_values('Email').drop_duplicates()    
```

```python
rock_list.reset_index().drop('index', axis=1).head()
```

### Question5: find the most popular music genre for each country
i.e genre with the most purchases

```python
# the description shows that ony one of each track is bought at a particular unitprice
genre_merge.describe()
```

```python
#group data by country and genre name and then sum the price to know the total price each country spent on a particular genre
country_genre = pd.DataFrame(genre_merge.groupby(['BillingCountry', 'Name'], as_index=False)['UnitPrice'].sum()) 
```

```python
country_genre.head(10)
```

```python
#group by country to know the maximum amount spent by countries
country_purchase = pd.DataFrame(country_genre.groupby('BillingCountry', as_index=False)['UnitPrice'].max())
```

```python
country_purchase.head()
```

```python
#join both dataframes together to see the countries and their most popular genres
pd.merge(country_purchase, country_genre, on = ['BillingCountry', 'UnitPrice'], how='inner')  
```
